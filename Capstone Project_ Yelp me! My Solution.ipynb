# Hypothesis 1: Restaurants that are open have better ratings that those that are not

```python
#Two sample test with a 1 tail
```

```python
#Alternative Hypothesis: Open restaurant ratings > closed restaurant ratings
#Null Hypothesis: Open restaurant ratings <= closed restaurant ratings
#Compare the length of the data
open_rest_stars = restaurant_df[restaurant_df['is_open']==1]['stars']
closed_rest_stars = restaurant_df[restaurant_df['is_open']==0]['stars']
```

```python
#Look at the mean
print(f"The start rating mean for open restaurants is {open_rest_stars.mean()}")
print(f"The start rating mean for closed restaurants is {closed_rest_stars.mean()}")
```

```python
#Create a function to read the p-value
def p_value_reader(p_value, alpha):
  if p_value < alpha:
    print("Reject the Null Hypothesis")
  else:
    print("Fail to reject the Null Hypothesis")


def test_2sample(sample1, sample2, alpha, alternative):
  #Levene's test
  stat, p_value = st.levene(sample1, sample2)
  #Interpret the test
  if p_value < alpha:
    equal_var = False
    print("Reject the Null Hypothesis. Variances are unequal. Perform Welch's Test")
  else:
    equal_var = True
    print("Fail to reject the Null hypothesis. Variances are equal. Perform 2-sample T-test")
  #2 sample test
  t_statistic, p_value = st.ttest_ind(sample1,
                                      sample2,
                                      equal_var = equal_var,
                                      alternative = alternative)
  print(f"The p-value is {p_value}")
  p_value_reader(p_value, alpha)
```

```python
#Apply the function
test_2sample(open_rest_stars, closed_rest_stars, alpha=0.05, alternative='greater')
```

# Hypothesis 2: Restaurants that deliver food have worse ratings

```python
# Null Hypothesis: Restaurants that delivery food ratings >= restaurants that don't deliver food ratings
# Alternative Hypothesis: Restaurants that delivery food ratings < restaurants that don't deliver food ratings


#What kind of test? two-sample test with 1 tail

#Create a copy of the df
df_h2 = restaurant_df.copy()
```

```python
#define a function to extract the 'RestaurantsDelivery' value
def is_delivery(attributes):
  if attributes and 'RestaurantsDelivery' in attributes:
    return attributes['RestaurantsDelivery'] == 'True'
  return False

#Apply the function to the 'attributes' column and create a new column 'Delivers food'
df_h2['delivers_food'] = df_h2['attributes'].apply(is_delivery)

# #Now, the 'delivers_food' column will contain True if the restaurant delivers, and False if it doesnt
# df.head()

```

```python
# Star rating for delivery and non-delivery restaurants
delivery_stars = df_h2[df_h2['delivers_food'] == True]['stars']
non_delivery_stars = df_h2[df_h2['delivers_food'] == False]['stars']
```

```python
#Look at the mean
print(f"The start rating mean for delivery restaurants is {delivery_stars.mean()}")
print(f"The start rating mean for non-delivery restaurants is {non_delivery_stars.mean()}")
```

```python
#Apply the function
test_2sample(delivery_stars, non_delivery_stars, alpha=0.05, alternative='less')
```

# Hypothesis 3: Restaurants that allow smoking are less likely to be open

```python
#Null Hypothesis: There's no relationship between teh variables (smoking and being open)
#Alternative Hypothesis: Restaurants that allow smoking are less likely to be open

#Test: Chisquare test
```

```python
df_h3 = restaurant_df.copy()
```

```python
#Create a contigency table to observe the frequency distribution between the smoking
contingency_table = pd.crosstab(df_h3['allows_smoking'], df_h3['is_open'])
print(contingency_table)
```
